```python
import pandas as pd

file_path = "/content/drive/MyDrive/week_2/finetuning/processed_jeju.csv"
data_df = pd.read_csv(file_path, index_col=0, header=0)
data_df
```

```python
df_train = data_df.sample(frac=0.8, random_state=1)
print(df_train)
print()
print(f"length of train df: {len(df_train)}")
```

```python
df_remain = data_df.drop(df_train.index)
df_eval = df_remain.sample(frac=0.5, random_state=1)
df_test = df_remain.drop(df_eval.index)
print(df_eval)
print()
print(f"length of test df: {len(df_eval)}")
print()
print(df_test)
print()
print(f"length of test df: {len(df_test)}")
```

```python
train_path = "/content/drive/MyDrive/week_2/finetuning/processed_jeju_train.csv"
eval_path = "/content/drive/MyDrive/week_2/finetuning/processed_jeju_eval.csv"
test_path = "/content/drive/MyDrive/week_2/finetuning/processed_jeju_test.csv"
df_train.to_csv(train_path)
df_eval.to_csv(eval_path)
df_test.to_csv(test_path)
```
