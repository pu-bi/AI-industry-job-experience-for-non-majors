###re.match

re.matchëŠ” ë¬¸ìì—´ì˜ ì‹œì‘ë¶€í„° ì •ê·œì‹ê³¼ ì¼ì¹˜í•˜ëŠ” ë¶€ë¶„ì„ ì°¾ì•„ëƒ…ë‹ˆë‹¤. ê²°ê³¼ë¡œ Match objectë¥¼ ë°˜í™˜í•˜ê¸° ë•Œë¬¸ì— ë¬¸ìì—´ë¡œ ì“¸ ë•ŒëŠ” group() ë©”ì†Œë“œë¥¼ ì¨ì¤ë‹ˆë‹¤.

###Grouping

ê·¸ë£¨í•‘ ë©”íƒ€ ê¸°í˜¸ ( )ë¥¼ ì˜ í™œìš©í•˜ë©´, match ê°ì²´ì˜ group ì¸ë±ìŠ¤ë¡œ ê·¸ë£¨í•‘ëœ ë¶€ë¶„ì˜ ë¬¸ìì—´ì„ ì¶”ì¶œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
print('without grouping')
p = re.compile("\w+\s[0-9]+")
m = p.findall("park 01012341234")
print(m)

print()
print('with grouping')
p = re.compile("(\w+)\s[0-9]")
m = p.findall("park 01012341234")
print(m)
````

````python
p = re.compile("(\w+)\s+(\d+[-]\d+[-]\d+)")
m = p.search("park    010-1234-1234")
print(m.group(0)) #ë§¤ì¹˜ëœ ë¬¸ì ì „ì²´
print(m.group(1)) #ì²«ë²ˆì§¸ ê´„í˜¸ ì•ˆì˜ ë¬¸ì
print(m.group(2)) #ë‘ë²ˆì§¸ ê´„í˜¸ ì•ˆì˜ ë¬¸ì
````

###Look-ahead

ì „ë°© íƒìƒ‰(look-ahead)ì€ ì°¾ì•„ë‚´ê³ ì í•˜ëŠ” ë¬¸ìì—´ì˜ íŒ¨í„´ì˜ ì „ë°©ì— íŠ¹ì • ì¡°ê±´ì„ ê±¸ê³  ì‹¶ì„ ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤. 

ì „ë°©ì— Xê°€ ë§¤ì¹˜ë˜ëŠ” ì¡°ê±´ì„ ê±¸ê³  ì‹¶ì„ ë•ŒëŠ” (?=X)

ì „ë°©ì— Xê°€ ë§¤ì¹˜ë˜ì§€ ì•ŠëŠ” ì¡°ê±´ì„ ê±¸ê³  ì‹¶ì„ ë•ŒëŠ” (?!X)

````python
p = re.compile(".+(?=:)")
result = p.search("http://google.com").group()
print(result)
````

````python
p = re.compile("[a-z]+(?![0-9])")
result = p.search("hello2022!").group()
print(result)
````

###Look-behind

í›„ë°© íƒìƒ‰(look-behind)ì€ ì°¾ì•„ë‚´ê³ ì í•˜ëŠ” ë¬¸ìì—´ì˜ íŒ¨í„´ì˜ í›„ë°©ì— íŠ¹ì • ì¡°ê±´ì„ ê±¸ê³  ì‹¶ì„ ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤. 

í›„ë°©ì— Xê°€ ë§¤ì¹˜ë˜ëŠ” ì¡°ê±´ì„ ê±¸ê³  ì‹¶ì„ ë•ŒëŠ” (?<=X)

í›„ë°©ì— Xê°€ ë§¤ì¹˜ë˜ì§€ ì•ŠëŠ” ì¡°ê±´ì„ ê±¸ê³  ì‹¶ì„ ë•ŒëŠ” (?<!X)

````python
p = re.compile("(?<=://).+")
result = p.search("http://google.com").group()
print(result)
````

````python
p = re.compile("(?<![a-z])\d+")
result = p.search("hello2022!").group()
print(result)
````
